```python
#each time starting session,get command ,run it, send output to mobile,then close the session and then start it again
import socket
import subprocess
while True:
    s=socket.socket()
    port=2222
    ip="192.168.99.1"
    s.bind((ip,port))
    s.listen()
    session,addr=s.accept()
    print(addr)
    command=session.recv(1000000000).decode()
    print(command)
    output=subprocess.getstatusoutput('{}'.format(command))
    #output = subprocess.run(["powershell", "-Command", command], capture_output=True)
    print(output[1])
    session.send(output[1].encode())
    print('sent')
    session.close()
```
